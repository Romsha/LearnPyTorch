```python
import torch
from torch import nn
from matplotlib import pyplot as plt
```

```python
weight = 0.7
bias = 0.3
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)
Y = X * weight + bias
```

```python
X[:10],Y[:10]
```

```python
train_split = int(0.8 * len(X))
X_train, Y_train = X[:train_split], Y[:train_split]
X_test, Y_test = X[train_split:], Y[train_split:]
```

```python
plt.plot(X_train, Y_train)
plt.plot(X_test, Y_test)
```

```python
def plot_pred(train_data=X_train, train_labaels=Y_train, test_data=X_test, test_labels=Y_test, predictions=None):
    plt.figure(figsize=(7, 5))
    plt.scatter(train_data, train_labaels, c="b", s=4, label="Training Data")
    plt.scatter(test_data, test_labels, c="r", s=4, label="Test Data")
    if predictions is not None:
        plt.scatter(test_data, predictions, c="g", s=4, label="Predicion")
    plt.legend(prop={'size': 10})
```

```python
plot_pred()
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float32))
        self.bias = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float32))

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weight * x + self.bias
```

```python
torch.manual_seed(42)
model_1 = LinearRegressionModel()
model_1.state_dict()
```

```python
with torch.inference_mode():
    y_preds = model_1(X_test)

plot_pred(predictions=y_preds)
```

```python
torch.manual_seed(42)
model_1 = LinearRegressionModel()
model_1.state_dict()
```

```python
loss_fn = nn.L1Loss()
optimizer = torch.optim.SGD(model_1.parameters(), lr=0.0001)
```

```python
losses = []
from IPython.display import clear_output
epochs = 20000
for epoch in range(epochs):
    model_1.train()
    
    y_pred = model_1(X_train)
    loss = loss_fn(y_pred, Y_train)
    losses.append(loss.item())
    # print(f'loss: {loss}')
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    model_1.eval()

with torch.inference_mode():
    plot_pred(predictions=model_1(X_test))
    plt.title(f'Epoch {epoch}, Loss: {loss}')
```

```python
def get_losses(lr, epochs=20000):
    torch.manual_seed(42)
    model_1 = LinearRegressionModel()
    model_1.state_dict()
    loss_fn = nn.L1Loss()
    optimizer = torch.optim.SGD(model_1.parameters(), lr=lr)
    losses = []
    
    for epoch in range(epochs):
        model_1.train()
        
        y_pred = model_1(X_train)
        loss = loss_fn(y_pred, Y_train)
        losses.append(loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
        model_1.eval()

    return losses
    
```

```python
lr_1 = get_losses(0.1, epochs=500)
lr_01 = get_losses(0.01, epochs=500)
lr_001 = get_losses(0.001, epochs=500)
lr_0001 = get_losses(0.0001, epochs=500)

plt.scatter(x=range(len(lr_0001)), y=lr_0001, s=1, label="lr=0.0001")
plt.scatter(x=range(len(lr_001)), y=lr_001, s=1, label="lr=0.001")
plt.scatter(x=range(len(lr_01)), y=lr_01, s=1, label="lr=0.01")
plt.scatter(x=range(len(lr_1)), y=lr_1, s=1, label="lr=0.1")
plt.legend(markerscale=5)
```

```python
lr_1 = get_losses(0.1)
lr_01 = get_losses(0.01)
lr_001 = get_losses(0.001)
lr_0001 = get_losses(0.0001)

plt.scatter(x=range(len(lr_1)), y=lr_1, s=1, label="lr=0.1")
plt.scatter(x=range(len(lr_01)), y=lr_01, s=1, label="lr=0.01")
plt.scatter(x=range(len(lr_001)), y=lr_001, s=1, label="lr=0.001")
plt.scatter(x=range(len(lr_0001)), y=lr_0001, s=1, label="lr=0.0001")
plt.legend(markerscale=5)
```
